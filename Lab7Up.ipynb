```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.visited = []

    def add_edge(self, u, v):
        self.graph[u].append(v)
        self.graph[v].append(u)

    def DFS(self, v, visited, path):
        visited[v] = True
        path.append(v)

        if len(path) == len(self.graph):
            return True

        for i in self.graph[v]:
            if not visited[i]:
                if self.DFS(i, visited, path):
                    return True

        visited[v] = False
        path.pop()
        return False

    def findHamiltonianPath(self):
        n = len(self.graph)
        if n < 2:
            return False
        
        for i in range(n):
            if len(self.graph[i]) < 1:
                return False

        path = []
        self.visited = [False] * n

        if self.DFS(0, self.visited, path):
            return path

        return False

    def findHamiltonianCircuit(self):
        n = len(self.graph)
        if n < 3:
            return False
        
        for i in range(n):
            if len(self.graph[i]) < 2:
                return False

        path = []
        self.visited = [False] * n

        if self.DFS(0, self.visited, path):
            if path[0] in self.graph[path[-1]]:
                path.append(path[0])
                return path

        return False

# Use Case
g = Graph()




g.add_edge(0,5)
g.add_edge(0,1)
g.add_edge(0,2)
g.add_edge(1,2)
g.add_edge(1,5)
g.add_edge(1,4)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(4,5)

hp = g.findHamiltonianPath()
if hp:
    print("There is a Hamiltonian Path and path is \n")
    print("Hamiltonian Path: ", hp)
else:
    print("No Hamiltonian Path exists.")

hc = g.findHamiltonianCircuit()
if hc:
    print("There is a Hamiltonian Circuit and path is \n")
    print("Hamiltonian Circuit: ", hc)
else:
    print("No Hamiltonian Circuit exists.")
```
